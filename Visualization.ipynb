#**Numpy**

#**Create array**

```python
import numpy as np
a=np.array([5,3,6,8,7])
b=np.arange(5,6)#creates[5,3,6,8,7]
c=np.zeros((3,6))#2x3 matrix of zeros
d=np.ones((3,5))#3x2matrix of ones
print("Array a:",a)
print("Array b:",b)
print("2x3 Zeros:",c)
print("3x2 Ones:",d)



```

#**Indexing And Slicing**

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2])
```

#**Basic arithmetic operations**

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

#**Basic Statitics**

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard deviation:",np.std(a))
```

#**Reshapping arrays**

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix")
matrix
```

#**Transpose**

```python
print("Transpose:")
matrix.T
```

#**Boolean Indexing**

```python
print("Element > 5:",matrix[matrix > 5])
```

#**Pandas**

#**Create DataFrame manually**

```python
import pandas as pd
data={
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

#**Basic Operations**

```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

#**Filtering**

```python
print("people with salary > 50000:")
df[df['Salary'] > 50000]
```

#**Sorting**

```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

#**Grouping And Aggregation**

```python
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary By Department")
grouped
```

#**Adding New columns**

```python
df['Salary in Lakhs'] = df['Salary'] /100000
df
```

#**MatPlotLib:Building Blocks Of Visualization**

#**LinePlot**

```python
import matplotlib.pyplot as plt
years=[2019,2020,2021,2022]
revenue=[100,120,130,160]

plt.plot(years,revenue,marker='o',linestyle='--',color='brown')
plt.title("Company Revenue Over Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in Crores)")
plt.xticks(years)
plt.grid(False)
plt.show()
```

#**Bar Chart**

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='blue')
plt.title("Employee Count By Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

#**Histogram**

```python
import matplotlib.pyplot as plt
ages=df['Age']
plt.hist(ages,bins=5,color='pink',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#**Pie Chart**

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proporation")
plt.ylabel("")
plt.show()
```

#**SeaBorn:Statistical Data Visualization**

```python
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()
print("\nDescription  of the tips datasetL:")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Count of Records by Day")
plt.show()
```

#**BoxPlot**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
